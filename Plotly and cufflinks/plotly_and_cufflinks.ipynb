```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
%matplotlib inline
```

```python
init_notebook_mode(connected=True)    # sets connection of plotly with pandas
cf.go_offline()    # online versions are paid
```

```python
df = pd.DataFrame(np.random.randn(100,4), columns='A B C D'.split())
df.head()
```

```python
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [32, 43, 50]})
df2
```

```python
df.plot()
```

```python
df.iplot()
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers', size=20)
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.sum().iplot(kind='bar')    # as in df there are no cathegorical columns, you need to use an aggregate funtion to use bar plot
```

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1,2,3,4,5], 'y': [10,20,30,20,10], 'z': [5,4,3,2,1]})
df3
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')    # 3D plot
```

```python
df['A'].iplot(kind='hist', bins=50)
```

```python
df.iplot(kind='hist')
```

```python
df[['A', 'B']].iplot(kind='spread')    # often used dor stocks data
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')    # often used for GDP, happiness etc. measuring
```

```python
df.scatter_matrix()    # similar to seaborn's pairplot
```
