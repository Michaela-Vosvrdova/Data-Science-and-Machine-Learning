```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x, y, 'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r')

plt.subplot(1,2,2)
plt.plot(y, x, 'b')
```

```python
# Object-oriented
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('LARGER PLOT')

axes2.plot(y, x, 'r')
axes2.set_title('SMALLER PLOT')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
```

## Figure size and DPI

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi=300)    # dpi = resolution (rozlišení)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='X Squared')
ax.plot(x, x**3, label='X Cubed')

ax.legend(loc=9)    # location of the legend, codes to be found in Matplotlib docs
```

## Plot Appearence

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='#FF8C00', linewidth=1, alpha=0.5, linestyle='-.', marker='o', markersize=20,
         markerfacecolor='brown', markeredgewidth=3, markeredgecolor='green')    # RGB Hex Code, 1 is the basic width of the line, alpha=transparency
# can use: linewidth/lw, linestyle/ls
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y, color='purple', lw=2, ls='--')
```
