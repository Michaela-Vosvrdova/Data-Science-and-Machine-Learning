## Categorical plots

```python
import seaborn as sns
import numpy as np
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# barplot: x = categorical value, y = numeric value; mean() is is used as an estimator in the bars
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
# creating own estimator
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# counting rows (records) per category, that's why we need just x -> there are more records of men than women
sns.countplot(x='sex', data=tips)
```

```python
# counting rows (same as count() in pandas)
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
# combination of stripplot and violinplot, don't use for the very large datasets
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='Black')
```

```python
# can call any kind of plot generally (bar, violin, ...)
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```
